# 4.Data Description

## 1. Date

### The calendar date on which the booking was created.

## 2. Time

## The exact time of the booking request.

## 3. Booking ID

### A unique identifier assigned to each ride booking.

## 4. Booking Status

### Final status of the booking, including:

#### Completed

#### Incomplete

#### Cancelled by Customer

#### Cancelled by Driver

#### No Driver Found

## 5. Customer ID

### A unique number identifying each customer.

## 6. Vehicle Type

### Type of vehicle selected by the customer, such as:

#### Bike

#### E-Bike

#### Go Mini

#### Go Sedan

#### Premier Sedan

#### Uber XL

## 7. Pickup Location

### Starting point of the ride, mostly within Delhi–NCR regions (Delhi, Noida, Gurugram, Ghaziabad, Faridabad).

## 8. Drop Location

### Destination location for the ride.

## 9. Avg VTAT (Vehicle Turnaround Time)

### Average time a driver takes to become available after completing a trip.

## 10. Avg CTAT (Customer Turnaround Time)

### Average time a customer takes between booking consecutive rides.

## 11. Cancelled Rides by Customer

### Indicates how many times the customer cancelled this booking.

## 12. Reason for Cancelling by Customer

### Reasons provided by customers, including:

#### AC not working

#### Change of plans

#### Driver asked to cancel

#### Driver not moving toward pickup

#### Wrong address

#### Long wait time

#### Driver unreachable

## 13. Cancelled Rides by Driver

### Indicates driver-initiated cancellations.

## 14. Driver Cancellation Reason

### Reasons given by drivers, including:

#### Customer-related issue

#### More passengers than allowed

#### Personal or car-related issues

#### Customer was sick

#### Unsafe/too far pickup

## 15. Incomplete Rides

### Shows whether the ride started but did not complete.

## 16. Incomplete Rides Reason

### Reasons for incomplete trips, such as:

#### Customer demand

#### Vehicle issue

#### Other issues

#### App or payment interruptions

## 17. Booking Value

### Total fare charged for the ride.

## 18. Ride Distance

### Distance traveled during the trip in kilometers.

## 19. Driver Ratings

### Rating provided by the customer to the driver (1–5).

## 20. Customer Rating

### Rating provided by the driver to the customer (1–5).

## 21. Payment Method

### Mode of payment used for the ride:

#### Cash

#### Credit Card

#### Debit Card

#### Uber Wallet

#### UPI

```python
df['Cancelled Rides by Customer'] = df['Cancelled Rides by Customer'].apply(
    lambda x: 'Customer' if x == 1 else 'Non-Cancellation'
)
```

```python
df['Cancelled Rides by Driver'] = df['Cancelled Rides by Driver'].apply(
    lambda x: 'Driver' if x == 1 else 'Non-Cancellation'
)
```

```python
df.isnull().sum()
```

```python
df.head(25)
```

```python
def merge_cancel(row):
    cust = str(row['Cancelled Rides by Customer']).lower()
    driv = str(row['Cancelled Rides by Driver']).lower()

    if ('non' in cust) and ('non' in driv):
        return 'Non-Cancellation'
    if 'customer' in cust:
        return 'Customer'
    if 'driver' in driv:
        return 'Driver'
    return 'Non-Cancellation'

df['Cancelled By'] = df.apply(merge_cancel, axis=1)
```

```python
df.head(25)
```

```python
# Get all columns
cols = list(df.columns)

# Remove Cancelled By from the list temporarily
cols.remove('Cancelled By')

# Find index of Avg CTAT (use exact name from your dataset)
idx = cols.index('Avg CTAT') + 1

# Insert Cancelled By after Avg CTAT
cols.insert(idx, 'Cancelled By')

# Reorder dataframe
df = df[cols]
```

```python
## DEMAND — Are people booking rides?

### When do bookings peak?

### Is demand consistent across days?
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bookings by hour
hourly_bookings = df.groupby('hour')['Booking ID'].count()

plt.figure(figsize=(8,5))
sns.lineplot(x=hourly_bookings.index, y=hourly_bookings.values)
plt.title('Demand Pattern: Bookings by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Bookings')
plt.show()

# Bookings by weekday
weekday_bookings = df.groupby('weekday')['Booking ID'].count()

plt.figure(figsize=(8,5))
sns.barplot(x=weekday_bookings.index, y=weekday_bookings.values)
plt.title('Demand Pattern: Bookings by Weekday')
plt.xlabel('Weekday')
plt.ylabel('Number of Bookings')
plt.show()
```

1.Morning and evening are peak demand windows
2.Driver supply and surge strategies should focus on       these hours
3.Demand forecasting can be uniform across weekdays
4.No need for aggressive weekday-based pricing changes

```python
# Booking status distribution
status_counts = df['Booking Status'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=status_counts.index, y=status_counts.values)
plt.title('Ride Completion & Failure Breakdown')
plt.xlabel('Booking Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Customer vs Driver cancellations
cancel_compare = df['Booking Status'].value_counts()[[
    'Cancelled by Customer', 'Cancelled by Driver'
]]

plt.figure(figsize=(6,4))
sns.barplot(x=cancel_compare.index, y=cancel_compare.values)
plt.title('Customer vs Driver Cancellations')
plt.ylabel('Number of Cancellations')
plt.show()
```

```python
import pandas as pd

vehicle_cancel_summary = pd.DataFrame({
    'Customer Cancelled': df[df['Booking Status'] == 'Cancelled by Customer']
        .groupby('Vehicle Type')['Booking ID'].count(),
        
    'Driver Cancelled': df[df['Booking Status'] == 'Cancelled by Driver']
        .groupby('Vehicle Type')['Booking ID'].count()
}).fillna(0)

vehicle_cancel_summary['Total Cancellations'] = (
    vehicle_cancel_summary['Customer Cancelled'] +
    vehicle_cancel_summary['Driver Cancelled']
)

vehicle_cancel_summary
```

5.Optimization efforts should prioritize high-usage vehicle categories

```python
## REVENUE & LEAKAGE — Where is money lost?

### Which vehicles lose most revenue?

### When does revenue loss happen?
```

```python
cancel_df = df[df['Booking Status'] != 'Completed']

# Revenue lost by vehicle
rev_lost_vehicle = cancel_df.groupby('Vehicle Type')['Fare'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=rev_lost_vehicle.index, y=rev_lost_vehicle.values)
plt.title('Revenue Leakage by Vehicle Type')
plt.ylabel('Revenue Lost')
plt.xticks(rotation=45)
plt.show()

# Revenue lost by hour
rev_lost_hour = cancel_df.groupby('hour')['Fare'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=rev_lost_hour.index, y=rev_lost_hour.values)
plt.title('Revenue Leakage by Time of Day')
plt.xlabel('Hour')
plt.ylabel('Revenue Lost')
plt.show()
```

```python
## PERFORMANCE — Which vehicles perform better?

### Which vehicles complete more rides?

### Which generate more revenue?
```

```python
# Completion rate by vehicle
vehicle_total = df.groupby('Vehicle Type')['Booking ID'].count()
vehicle_completed = df[df['Booking Status']=='Completed'].groupby('Vehicle Type')['Booking ID'].count()
completion_rate = (vehicle_completed / vehicle_total * 100)

plt.figure(figsize=(8,5))
sns.barplot(x=completion_rate.index, y=completion_rate.values)
plt.title('Completion Rate by Vehicle Type')
plt.ylabel('Completion Rate (%)')
plt.xticks(rotation=45)
plt.show()

# Revenue by vehicle
rev_vehicle = df[df['Booking Status']=='Completed'].groupby('Vehicle Type')['Fare'].sum()

plt.figure(figsize=(8,5))
sns.barplot(x=rev_vehicle.index, y=rev_vehicle.values)
plt.title('Revenue by Vehicle Type')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```

1.Revenue loss is driven more by volume than by fare size
2.Improving reliability in high-demand vehicle categories   will have the greatest financial impact
3.Peak hours are both the most profitable and most         vulnerable periods
4.Insufficient driver supply during peak times directly translates into higher revenue loss

## BEHAVIOR — How do customers behave?

### How do users pay?

### Which method brings revenue?

```python
# Revenue by payment
rev_payment = df[df['Booking Status']=='Completed'].groupby('Payment Method')['Fare'].sum()

plt.figure(figsize=(6,6))
plt.pie(rev_payment.values, labels=rev_payment.index, autopct='%1.1f%%')
plt.title('Revenue Share by Payment Method')
plt.show()

# Avg fare by payment
avg_fare_payment = df[df['Booking Status']=='Completed'].groupby('Payment Method')['Fare'].mean()

plt.figure(figsize=(8,5))
sns.barplot(x=avg_fare_payment.index, y=avg_fare_payment.values)
plt.title('Average Fare by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

1.UPI is a critical revenue driver
2.Maintaining seamless UPI payment experiences is           essential for revenue stability
3.Revenue differences across payment methods are driven     by usage volume, not higher fares
4.Incentivizing preferred payment methods can increase     revenue without impacting fare levels

```python
## LOCATION & ROUTES — Where should Uber focus?
## Which routes matter most?
```

```python
df['Route'] = df['Pickup Location'] + ' → ' + df['Drop Location']

# Top routes by volume
route_volume = df[df['Booking Status']=='Completed'].groupby('Route')['Booking ID'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=route_volume.index, y=route_volume.values)
plt.title('Top Pickup–Drop Routes by Volume')
plt.xticks(rotation=45, ha='right')
plt.show()

# Top routes by revenue
route_revenue = df[df['Booking Status']=='Completed'].groupby('Route')['Fare'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=route_revenue.index, y=route_revenue.values)
plt.title('Top Pickup–Drop Routes by Revenue')
plt.xticks(rotation=45, ha='right')
plt.show()
```

1.High-volume routes are critical for service reliability
2.Ensuring driver availability on these routes can         prevent delays and cancellations
3.Revenue concentration is driven by both trip length and   pricing, not just frequency
4.These routes are strong candidates for targeted surge pricing or premium vehicle allocation
